TASK 1   
Consider a standard deck of 52 cards. What is the probability of drawing a red card (hearts or diamonds)?    
Now, given that you have drawn a red card, what is the probability that the card is a heart?                    
Given that the drawn card is a face card (Jack,Queen, King), what is the probability that it is a diamond?      
Given that the card drawn is a face card, calculate the probability that it is also a spade or a queen.                

ANSWER:      
i)   P(Red) = 26/52 = 1/2 = 0.50              
ii)  P(Heart | Red) = 13/26 = 1/2 = 0.50          
iii) P(Diamond | Face) = 3/12 = 1/4 = 0.25                              
iv)  P(Spade OR Queen | Face) = (3 + 4 − 1) / 12 = 6/12 = 1/2 = 0.50    


Task 2
You are given the following Bayesian Network structure to model student exam performance:   
Nodes:                                
● Intelligence (I) — {High, Low}                    
● StudyHours (S) — {Sufficient, Insufficient}      
● Difficulty (D) — {Hard, Easy}         
● Grade (G) — {A, B, C}            
● Pass (P) — {Yes, No}       
Dependencies:             
Intelligence, StudyHours, and Difficulty directly affect Grade.                                   
Grade directly affects Pass.         
Prior Probabilities:                                        
● P(Intelligence = High) = 0.7, P(Intelligence = Low) = 0.3  
● P(StudyHours = Sufficient) = 0.6, P(StudyHours = Insufficient) = 0.4               
● P(Difficulty = Hard) = 0.4, P(Difficulty = Easy) = 0.6    
Conditional Probabilities (examples):               
P(Grade | Intelligence, StudyHours, Difficulty):   
(Assume your own valid values, e.g., students with High intelligence, Sufficientstudy hours, and Easy difficulty are most likely to get A)
● P(Pass | Grade):                       
● P(Pass = Yes | Grade = A) = 0.95       
● P(Pass = Yes | Grade = B) = 0.80          
● P(Pass = Yes | Grade = C) = 0.50       
Tasks to do         
● Construct the Bayesian Network structure diagram showing all dependencies.                       
● Define the complete Conditional Probability Tables (CPTs) for all nodes.              
● Implement the Bayesian Network using Python (pgmpy or equivalent).               
● Perform inference using Variable Elimination to answer:
● What is the probability that the student passes the exam, given:               
StudyHours = Sufficient, Difficulty = Hard
● What is the probability that the student has High Intelligence, given:           
Pass = Yes  

```python
import subprocess, sys
subprocess.run([sys.executable, "-m", "pip", "install", "pgmpy", "-q","--break-system-packages"], capture_output=True)

from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

model_student = DiscreteBayesianNetwork([
    ('Intelligence', 'Grade'),
    ('StudyHours',   'Grade'),
    ('Difficulty',   'Grade'),
    ('Grade',        'Pass')
])


cpd_intel = TabularCPD('Intelligence', 2, values=[[0.7], [0.3]])


cpd_study = TabularCPD('StudyHours', 2, values=[[0.6], [0.4]])

cpd_diff = TabularCPD('Difficulty', 2, values=[[0.4], [0.6]])

cpd_grade = TabularCPD(
    variable='Grade',
    variable_card=3,
    values=[
      [0.60, 0.85, 0.30, 0.55, 0.25, 0.50, 0.10, 0.20],
      [0.30, 0.12, 0.45, 0.35, 0.45, 0.35, 0.35, 0.45],
      [0.10, 0.03, 0.25, 0.10, 0.30, 0.15, 0.55, 0.35],
    ],
    evidence=['Intelligence', 'StudyHours', 'Difficulty'],
    evidence_card=[2, 2, 2]
)


cpd_pass = TabularCPD(
    variable='Pass',
    variable_card=2,
    values=[
        [0.95, 0.80, 0.50],   
        [0.05, 0.20, 0.50],   
    ],
    evidence=['Grade'],
    evidence_card=[3]
)

model_student.add_cpds(cpd_intel, cpd_study, cpd_diff, cpd_grade, cpd_pass)
assert model_student.check_model(), "Student model check failed!"
print("\n Student Bayesian Network model verified successfully.")

infer_student = VariableElimination(model_student)

q1 = infer_student.query(
    variables=['Pass'],
    evidence={'StudyHours': 0, 'Difficulty': 0},
    show_progress=False
)
print("\nQuery 1: P(Pass | StudyHours=Sufficient, Difficulty=Hard)")
print(q1)
print(f"P(Pass=Yes) = {q1.values[0]:.4f}")
print(f"P(Pass=No)  = {q1.values[1]:.4f}")

q2 = infer_student.query(
    variables=['Intelligence'],
    evidence={'Pass': 0},
    show_progress=False
)
print("\nQuery 2: P(Intelligence | Pass=Yes)")
print(q2)
print(f"P(Intelligence=High | Pass=Yes) = {q2.values[0]:.4f}")
print(f"P(Intelligence=Low  | Pass=Yes) = {q2.values[1]:.4f}")
```

Task 3    
You are tasked with building a Bayesian Network to predict the likelihood of a disease (e.g., Flu or Cold) based on the presence of various symptoms, including fever, cough, fatigue, and chills. The network consists of the following nodes:          
● Symptoms: Fever, Cough, Fatigue, Chills       
● Disease: Flu, Cold       
Network Structure:         
The Disease node influences the symptoms: Fever, Cough, Fatigue, and Chills.           
The Symptoms nodes are observed, and they are conditionally dependent on the Disease node.   

Prior Probabilities:     

P(Disease):            
1. P(Flu) = 0.3         
2. P(Cold) = 0.7         

P(Symptoms | Disease): (You are to define the following conditional probabilities based on assumptions )  

P(Fever | Disease):                 
1. P(Fever = Yes | Flu) = 0.9          
2. P(Fever = Yes | Cold) = 0.5       
3. P(Fever = No | Flu) = 0.1         
4. P(Fever = No | Cold) = 0.5       
 
P(Cough | Disease):                
1. P(Cough = Yes | Flu) = 0.8                
2. P(Cough = Yes | Cold) = 0.6          
3. P(Cough = No | Flu) = 0.2               
4. P(Cough = No | Cold) = 0.4           
                         
P(Fatigue | Disease):                 
1. P(Fatigue = Yes | Flu) = 0.7           
2. P(Fatigue = Yes | Cold) = 0.3         
3. P(Fatigue = No | Flu) = 0.3         
4. P(Fatigue = No | Cold) = 0.7      
                           
P(Chills | Disease):                
1. P(Chills = Yes | Flu) = 0.6           
2. P(Chills = Yes | Cold) = 0.4         
3. P(Chills = No | Flu) = 0.4          
4. P(Chills = No | Cold) = 0.6           

Construct the Bayesian Network:             
Draw the structure of the Bayesian Network based on the dependencies given above.                           
Define the Conditional Probability Tables (CPTs)    
Define reasonable CPTs for all nodes (Disease, Fever, Cough, Fatigue, and Chills) based on the given probabilities. You may assume the probabilities if necessary.     

Perform Inference Task 1:              
Use the Bayesian Network to calculate the posterior probability of the Disease (Flu or Cold), given the following symptoms:            

● Fever = Yes         
● Cough = Yes          

Perform Inference Task 2:      

Add a new symptom: Chills.      

Update the Bayesian Network to compute the posterior probability of the Disease (Flu or Cold), given the following symptoms:                
● Fever = Yes         
● Cough = Yes           
● Chills = Yes        

Perform Inference Task 3:       

Given that the disease is Flu, calculate the probability that the person also has Fatigue.          
● What is P(Fatigue = Yes | Disease = Flu)?



```python
model_disease = DiscreteBayesianNetwork([
    ('Disease', 'Fever'),
    ('Disease', 'Cough'),
    ('Disease', 'Fatigue'),
    ('Disease', 'Chills'),
])

cpd_disease = TabularCPD('Disease', 2, values=[[0.3], [0.7]])

cpd_fever = TabularCPD(
    'Fever', 2,
    values=[[0.9, 0.5],  
            [0.1, 0.5]],  
    evidence=['Disease'], evidence_card=[2]
)

cpd_cough = TabularCPD(
    'Cough', 2,
    values=[[0.8, 0.6],
            [0.2, 0.4]],
    evidence=['Disease'], evidence_card=[2]
)

cpd_fatigue = TabularCPD(
    'Fatigue', 2,
    values=[[0.7, 0.3],
            [0.3, 0.7]],
    evidence=['Disease'], evidence_card=[2]
)

cpd_chills = TabularCPD(
    'Chills', 2,
    values=[[0.6, 0.4],
            [0.4, 0.6]],
    evidence=['Disease'], evidence_card=[2]
)

model_disease.add_cpds(cpd_disease, cpd_fever, cpd_cough, cpd_fatigue, cpd_chills)
assert model_disease.check_model(), "Disease model check failed!"
print("\nDisease Bayesian Network model verified successfully.")

infer_disease = VariableElimination(model_disease)

it1 = infer_disease.query(
    variables=['Disease'],
    evidence={'Fever': 0, 'Cough': 0},   # 0=Yes
    show_progress=False
)
print("\nInference Task 1: P(Disease | Fever=Yes, Cough=Yes)")
print(it1)
print(f"P(Flu  | Fever=Yes, Cough=Yes) = {it1.values[0]:.4f}")
print(f"P(Cold | Fever=Yes, Cough=Yes) = {it1.values[1]:.4f}")


it2 = infer_disease.query(
    variables=['Disease'],
    evidence={'Fever': 0, 'Cough': 0, 'Chills': 0},
    show_progress=False
)
print("\nInference Task 2: P(Disease | Fever=Yes, Cough=Yes, Chills=Yes)")
print(it2)
print(f"P(Flu  | Fever, Cough, Chills = Yes) = {it2.values[0]:.4f}")
print(f"P(Cold | Fever, Cough, Chills = Yes) = {it2.values[1]:.4f}")

p_fatigue_flu = 0.7
print(f"\nInference Task 3: P(Fatigue=Yes | Disease=Flu)")
print(f"This is directly read from the CPT: {p_fatigue_flu}")

it3 = infer_disease.query(
    variables=['Fatigue'],
    evidence={'Disease': 0},  
    show_progress=False
)
print(f"Confirmed via Variable Elimination: P(Fatigue=Yes | Flu) = {it3.values[0]:.4f}")
```

TASK 4         
Define a simple Markov Model with three weather states: Sunny, Cloudy, and Rainy.          
Create a transition matrix for the weather states, specifying the probabilities of transitioning from one state to another (e.g., from Sunny to Cloudy, or from Cloudy to Rainy).   Simulate the weather for the next 10 days, starting with a Sunny day, using the Markov Model.     
Calculate the probability of having at least 3 rainy days over the 10-day period. 

```python
import numpy as np

states = ["Sunny", "Cloudy", "Rainy"]
state_index = {s: i for i, s in enumerate(states)}

transition_matrix = np.array([
    [  0.60,  0.30,   0.10],   
    [  0.30,  0.40,   0.30],   
    [  0.20,  0.30,   0.50],   
])

print("\nTransition Matrix:")
print(f"{'':10} {'Sunny':>8} {'Cloudy':>8} {'Rainy':>8}")
for i, row in enumerate(transition_matrix):
    print(f"{states[i]:10} {row[0]:>8.2f} {row[1]:>8.2f} {row[2]:>8.2f}")

def simulate_weather(initial_state, num_days, seed=None):
    if seed is not None:
        np.random.seed(seed)
    current = state_index[initial_state]
    sequence = [initial_state]
    for _ in range(num_days - 1):
        current = np.random.choice(len(states), p=transition_matrix[current])
        sequence.append(states[current])
    return sequence

np.random.seed(42)
weather_seq = simulate_weather("Sunny", 10, seed=42)
print(f"\n10-Day Simulation (starting Sunny):")
print(weather_seq)
rainy_count = weather_seq.count("Rainy")
print(f"Rainy days in simulation: {rainy_count}")

num_simulations = 100_000
rainy_threshold = 3
count_at_least_3_rainy = 0

for _ in range(num_simulations):
    seq = simulate_weather("Sunny", 10)
    if seq.count("Rainy") >= rainy_threshold:
        count_at_least_3_rainy += 1

p_at_least_3_rainy = count_at_least_3_rainy / num_simulations
print(f"\nMonte Carlo Estimate ({num_simulations:,} simulations):")
print(f"P(at least 3 Rainy days in 10 days | Start=Sunny)")
print(f"{p_at_least_3_rainy:.4f} ({p_at_least_3_rainy*100:.2f}%)")

from collections import Counter
rainy_dist = Counter()
for _ in range(num_simulations):
    seq = simulate_weather("Sunny", 10)
    rainy_dist[seq.count("Rainy")] += 1

print(f"\nDistribution of Rainy days over 10-day period:")
for k in sorted(rainy_dist):
    pct = rainy_dist[k] / num_simulations * 100
    bar = int(pct / 2)
    print(f"{k} rainy days: {pct:5.2f}% {bar}")
```
